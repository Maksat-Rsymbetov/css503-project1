```python
import requests
import math
import pandas as pd
from bs4 import BeautifulSoup as bs
from IPython.display import clear_output
import matplotlib.pyplot as plt
```

```python
url = 'https://www.lamoda.kz/c/1263/default-premium-men/?is_sale=1'
r = requests.get(url)
```

```python
soup = bs(r.text, 'html.parser')
product_card_dom_class = {'class': 'x-product-card__card'}
products_dom = soup.findAll('div', attrs=product_card_dom_class)
current_page = 2
# for i in range(2, 4):
while True:
    req = requests.get((url + '&page={page}').format(page=current_page))
    inner_soup = bs(req.text, 'html.parser')
    inner_dom = inner_soup.findAll('div', attrs=product_card_dom_class)
    if len(inner_dom) == 0: break
    products_dom.extend(inner_dom)
    current_page+=1

    clear_output()
    print('extending dom list: {page} {entries}'.format(page=current_page, entries=len(inner_dom)))

products = {
    'name': [],
    'brand': [],
    'old_price': [],
    'new_price': [],
    'discount': [],
}
index = 0
for i in products_dom:
    clear_output()
    print('processing information: {number}'.format(number=index))
    index+=1
    # Finding data in the DOM
    name = i.find('div', attrs={'class':'x-product-card-description__product-name'})
    brand = i.find('div', attrs={'class':'x-product-card-description__brand-name'})
    old_price_dom = i.find('span', attrs={'class':'x-product-card-description__price-old'})
    new_price_dom = i.find('span', attrs={'class':'x-product-card-description__price-new'})
    
    # Formatting and calculating data
    old_price = int(''.join(filter(str.isdigit, old_price_dom.text)))
    new_price = int(''.join(filter(str.isdigit, new_price_dom.text)))
    discount = 100 - ((new_price / old_price) * 100)

    # Adding to the dataset
    products['name'].append(name.text.strip())
    products['brand'].append(brand.text.strip())
    products['old_price'].append(old_price)
    products['new_price'].append(new_price)
    products['discount'].append(discount)
```

```python
dataframe = pd.DataFrame(products)
brand_price_mean = dataframe[['brand', 'new_price', 'old_price']].groupby(['brand']).mean(['new_price'])
brand_price_mean
```

```python
plt.figure(figsize=(10, 100), dpi=100)
plt.xlabel('Brand')
plt.ylabel('Price')
plt.title('Average price before and after discount')
plt.autoscale(tight=True)
ax = plt.gca()
brand_price_mean.sort_values(by=['new_price']).plot(grid=True, kind='barh', color=['r', 'b'], alpha=0.9, ax=ax)
plt.show()
```

```python
brand_discount_mean = dataframe[['brand', 'discount']].groupby(['brand']).mean(['discount'])
plt.figure(figsize=(10, 100), dpi=100)
plt.xlabel('Brand')
plt.ylabel('Discount (%)')
plt.title('Average discount for each brand product')
ax = plt.gca()
brand_discount_mean.sort_values(by=['discount']).plot(grid=True, kind='barh', color=['r', 'b'], alpha=0.9, ax=ax)
plt.show()
```

```python
price_correlation = dataframe[['old_price', 'new_price', 'discount']].sort_values(by=['new_price'])
price_correlation['count'] = range(0, len(price_correlation))
price_correlation['discount'] *= 10000
plt.figure(figsize=(50, 20), dpi=50)
plt.plot(price_correlation['count'], price_correlation['new_price'], label='Price after discount')
plt.plot(price_correlation['count'], price_correlation['old_price'], label='Price before discount')
plt.plot(price_correlation['count'], price_correlation['discount'], label='Discount')
plt.show()
```

```python
fpp = price_correlation
fpp['monetary_discount'] = fpp['old_price'] - fpp['new_price']
fpp = fpp.sort_values(by=['monetary_discount'])
fpp['count'] = range(0, len(fpp));

plt.figure(figsize=(100, 20), dpi=50);
plt.plot(fpp['count'], fpp['old_price'])
plt.plot(fpp['count'], fpp['new_price'])
plt.plot(fpp['count'], fpp['monetary_discount'])
plt.show()
```
